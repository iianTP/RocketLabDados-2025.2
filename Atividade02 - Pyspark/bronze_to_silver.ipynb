```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from itertools import chain

spark = SparkSession.builder.appName("Atividade2PraticaSpark").getOrCreate()

def rename_cols(df, cols):
    return df.select([F.col(old_name).alias(new_name) for old_name, new_name in cols.items()])
```

```python
ft_consumidores = spark.table("medalhao.bronze.ft_consumidores")

cols_ft_consumidores = {
    'customer_id':'id_consumidor',
    'customer_zip_code_prefix':'prefixo_cep',
    'customer_city':'cidade',
    'customer_state':'estado',
}

ft_consumidores_new = (
    rename_cols(ft_consumidores, cols_ft_consumidores)
    .withColumn('cidade',F.upper('cidade'))
    .withColumn('estado',F.upper('estado'))
    .dropDuplicates(['id_consumidor'])
)

ft_consumidores_new.select('*').display()

```

```python
ft_pedidos = spark.table("medalhao.bronze.ft_pedidos")

status = {
    'delivered': 'entregue',
    'invoiced': 'faturado',
    'shipped': 'enviado',
    'processing': 'em processamento',
    'unavailable': 'indisponvel',
    'canceled': 'cancelado',
    'created': 'criado',
    'approved': 'aprovado'
}

cols_ft_pedidos = {
    'order_id': 'id_pedido',
    'customer_id': 'id_consumidor',
    'order_status': 'status',
    'order_purchase_timestamp': 'pedido_compra_timestamp',
    'order_approved_at': 'pedido_aprovado_timestamp',
    'order_delivered_carrier_date': 'pedido_carregado_timestamp',
    'order_delivered_customer_date': 'pedido_entregue_timestamp',
    'order_estimated_delivery_date': 'pedido_estimativa_entrega_timestamp'
}

status_map = F.create_map([F.lit(s) for s in chain(*status.items())])

ft_pedidos_new = (
    rename_cols(ft_pedidos, cols_ft_pedidos)
    .withColumn('status',F.coalesce(status_map.getItem(F.col('status')),F.lit('n/a')))
    .withColumn('tempo_entrega_dias',
        F.datediff(F.col('pedido_entregue_timestamp'), F.col('pedido_compra_timestamp'))
    )
    .withColumn('tempo_entrega_estimado_dias',
        F.datediff(F.col('pedido_estimativa_entrega_timestamp'), F.col('pedido_compra_timestamp'))
    )
    .withColumn('diferenca_entrega_dias',
        F.col('tempo_entrega_dias')- F.col('tempo_entrega_estimado_dias')
    )
    .withColumn('entrega_no_prazo',
        F.when(F.col('diferenca_entrega_dias')<=0,F.lit('Sim'))
        .when(F.col('diferenca_entrega_dias')>0,F.lit('No'))
        .otherwise(F.lit('No Entregue'))
    )
)
#ft_pedidos_new.select('status').filter(F.col('status')=='n/a').display()
```

```python
ft_pagamentos = spark.table("medalhao.bronze.ft_pagamentos_pedidos")

cols_ft_pagamentos = {
    'order_id':'id_pedido',
    'payment_sequential':'codigo_pagamento',
    'payment_type':'forma_pagamento',
    'payment_installments':'parcelas',
    'payment_value':'valor_pagamento'
}
payment_type = {
    'credit_card':'Carto de Crdito',
    'boleto':'Boleto',
    'voucher':'Voucher',
    'debit_card':'Carto de Dbito'
}

payment_type_map = F.create_map([F.lit(p) for p in chain(*payment_type.items())])

ft_pagamentos_new = (
    rename_cols(ft_pagamentos, cols_ft_pagamentos)
    .withColumn('forma_pagamento',F.coalesce(payment_type_map.getItem(F.col('forma_pagamento')),F.lit('Outro')))
)

```

```python
"""
ID vlido:
- 32 caracteres
- somente caracteres hexadecimais (0-9, a-f)
- no pode ser nulo
- no pode ser vazio

Data vlida:
- formato yyyy-MM-dd HH:mm:ss
- no pode ser nulo
- no pode ser vazio
"""

ft_avaliacoes_pedidos = spark.table("medalhao.bronze.ft_avaliacoes_pedidos")

cols_ft_avaliacoes_pedidos = {
    'review_id': 'id_avaliacao',
    'order_id': 'id_pedido',
    'review_score': 'avaliacao',
    'review_comment_title': 'titulo_comentario',
    'review_comment_message': 'comentario',
    'review_creation_date': 'data_comentario',
    'review_answer_timestamp': 'data_resposta'
}

hash_regex = '^[0-9a-f]{32}$'

ft_avaliacoes_pedidos_new = (
    rename_cols(ft_avaliacoes_pedidos, cols_ft_avaliacoes_pedidos)
    .filter(
        F.col('review_id').rlike(hash_regex) & 
        F.isnotnull(F.try_to_timestamp(F.col('data_resposta'))) &
        F.isnotnull(F.try_to_timestamp(F.col('data_comentario')))
    )
    .filter(F.to_timestamp(F.col('data_resposta')) > F.to_timestamp(F.col('data_comentario')))
)

display(f'Linhas removidas: {ft_avaliacoes_pedidos.count()-ft_avaliacoes_pedidos_new.count()}') 
```

```python
ft_produtos = spark.table("medalhao.bronze.ft_produtos")
ft_vendedores = spark.table("medalhao.bronze.ft_vendedores")
ft_itens_pedidos = spark.table("medalhao.bronze.ft_itens_pedidos")
dm_categoria_produtos_traducao = spark.table("medalhao.bronze.dm_categoria_produtos_traducao")

cols_ft_itens_pedidos = {
    'order_id': 'id_pedido',
    'order_item_id': 'id_item',
    'product_id': 'id_produto',
    'seller_id': 'id_vendedor',
    'price': 'preco_BRL',
    'freight_value': 'preco_frete',
}

cols_ft_produtos = {
    'product_id': 'id_produto',
    'product_category_name': 'categoria_produto',
    'product_weight_g': 'peso_produto_gramas',
    'product_length_cm': 'comprimento_centimetros',
    'product_height_cm': 'altura_centimetros',
    'product_width_cm': 'largura_centimetros'
}

cols_ft_vendedores = {
    'seller_id': 'id_vendedor',
    'seller_zip_code_prefix': 'prefixo_cep',
    'seller_city': 'cidade',
    'seller_state': 'estado'
}

cols_dm_categoria_produtos_traducao = {
    'product_category_name': 'nome_produto_pt',
    'product_category_name_english': 'nome_produto_en'
}


ft_itens_pedidos_new = rename_cols(ft_itens_pedidos, cols_ft_itens_pedidos)
ft_produtos_new = rename_cols(ft_produtos, cols_ft_produtos)
dm_categoria_produtos_traducao_new = rename_cols(dm_categoria_produtos_traducao, cols_dm_categoria_produtos_traducao)
ft_vendedores_new = (
    rename_cols(ft_vendedores, cols_ft_vendedores)
    .withColumn('cidade',F.upper('cidade'))
    .withColumn('estado',F.upper('estado'))
)

```

```python
dm_cotacao_dolar = spark.table("medalhao.bronze.dm_cotacao_dolar")

cols_dm_cotacao_dolar = {
    'cotacaoCompra': 'cotacao_dolar',
    'dataHoraCotacao': 'data'
}

window = Window.orderBy("data")

dm_cotacao_dolar_renamed = rename_cols(dm_cotacao_dolar, cols_dm_cotacao_dolar)

min_date, max_date = (
    dm_cotacao_dolar_renamed
    .agg(
        F.min('data').alias('min_data'),
        F.max('data').alias('max_data')
    )
    .first()
)

df_calendario = (
    spark.range(1)
    .select(F.expr(
        f"sequence(to_date('{min_date}'), to_date('{max_date}'), interval 1 day)"
    )
    .alias("data_completa")) 
    .withColumn("data", F.expr("explode(data_completa)")) 
    .select("data")
)

dm_cotacao_dolar_nulos = df_calendario.join(dm_cotacao_dolar_renamed, on="data", how="left")

window_spec = Window.orderBy("data").rowsBetween(
    Window.unboundedPreceding, 
    Window.currentRow
)

# 5. Aplicar last_value(..., ignoreNulls=True)
dm_cotacao_dolar_new = (
    dm_cotacao_dolar_nulos.withColumn('cotacao_dolar_preenchida',
    F.last_value(F.col('cotacao_dolar'), True).over(window_spec))
    .select('data',F.col('cotacao_dolar_preenchida').alias('cotacao_dolar'))
)

```

```python
ft_consumidores = spark.table("medalhao.silver.ft_consumidores")

ft_pedidos = spark.table("medalhao.silver.ft_pedidos")

ft_pagamentos_pedidos = spark.table("medalhao.silver.ft_pagamentos_pedidos")

dm_cotacao_dolar = spark.table("medalhao.silver.dm_cotacao_dolar")

ft_pedido_total = (
    ft_consumidores
    .join(ft_pedidos, on='id_consumidor', how='left')
    .join(ft_pagamentos_pedidos, on='id_pedido', how='left')
    .join(dm_cotacao_dolar, 
            on=[
                ft_pedidos['pedido_compra_timestamp'].cast('date')==dm_cotacao_dolar['data']
            ], how='left'
    )
    .withColumn('valor_total_pago_brl',F.col('valor_pagamento'))
    .withColumn('valor_total_pago_usd',F.col('valor_pagamento')*F.col('cotacao_dolar'))
    .withColumn('data_pedido',F.col('pedido_compra_timestamp').cast('date'))
    .select('id_pedido','id_consumidor','status','valor_total_pago_brl',F.col('valor_total_pago_usd').cast('decimal(12,2)'),'data_pedido')
)

ft_pedido_total.write.format("delta").mode("overwrite").saveAsTable(f"medalhao.silver.ft_pedido_total")
```
